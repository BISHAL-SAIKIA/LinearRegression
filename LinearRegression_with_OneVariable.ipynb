```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('houseprice.csv')
df.head()
```

```python
%matplotlib inline
plt.title('HOUSE PRICE')
plt.xlabel('Area(sq feet)')
plt.ylabel('Price (US$)')
plt.scatter(df.area,df.price, color='orange', marker='+')
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']], df.price) ## first parameter- 2d array, second parameter 1d array
```

```python
reg.predict(pd.DataFrame([[3300]], columns=['area']))
```

```python
#y=mx+b
# for m:
reg.coef_
```

```python
#for b
reg.intercept_
```

```python
#using m and b values for verifying
135.78767123*3300 + 180616.43835616432
```

```python
#It mateches
```

```python
%matplotlib inline
plt.title('HOUSE PRICE')
plt.xlabel('Area(sq feet)')
plt.ylabel('Price (US$)')
plt.scatter(df.area,df.price, color='orange', marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
#As can be seen , the predicted line is denoted by the blue line which best decides the prediction model.
```

```python
#LETs say we have a csv with areas as column, we want to predict price for each and show it corresposnding to it
area_df=pd.read_csv('area.csv')
area_df.head()
```

```python
p= reg.predict(area_df)
area_df['prices']=p
area_df.sample()
```

```python
area_df.to_csv('prediction.csv', index=False)
```
